```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

data = {'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily', 'Frank', 'Gina', 'Saul', 'Paul', 'Nina'],
        'age': [25, 34, 18, 47, 29, 50, 27, 28, 33, 49],
        'gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'M', 'M', 'F'],
        'country': ['USA', 'Canada', 'USA', 'USA', 'Canada', 'USA', 'USA', 'USA', 'USA', 'Canada'],
        'salary': [50000, 65000, 35000, 80000, 45000, 70000, 55000, 45000, 70000, 85000],
        'department': ['Sales', 'Sales', 'Sales', 'Marketing', 'IT', 'Marketing', 'IT', 'Sales', 'Sales', 'Marketing'],
        'age_group': ["old", "old", "old", "old", "old", "old", "old", "old", "old", "old"],
        'age_range': [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan],
        'is_american': [True, False, True, False, True, True, True, True, True, True],
        'salary_rank': [4., 6., 1., 9., 2., 7., 5., 3., 8., 10.],
        'age_category': ['senior', 'senior', 'senior', 'senior', 'senior', 'senior', 'senior', 'senior', 'senior', 'senior'],
        'years_exp': [1, 1, 3, 8, 5, 2, 1, 2, 4, 7]
        }

df = pd.DataFrame(data)
df
```

```python
# 35. Create a 3D surface plot
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

# Reshape the data into a 2-dimensional array
age_unique = np.sort(df['age'].unique())
years_exp_unique = np.sort(df['years_exp'].unique())
age, years_exp = np.meshgrid(age_unique, years_exp_unique)
salary = df.pivot_table(index='age', columns='years_exp', values='salary').values

# Transpose the salary array to match the shape of age and years_exp
salary = np.transpose(salary)

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(age, years_exp, salary)

plt.show()
```

```python
# 46. Create a histogram with a density curve
import scipy.stats as stats

# Create histogram
plt.hist(df['age'], density=True, bins=10, alpha=0.5)

# Create density curve
mu, std = stats.norm.fit(df['age'])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Histogram of Age with Density Curve')

plt.show()
```

```python
# 47. Create a violin plot with separate colors for each group
fig, ax = plt.subplots()
colors = {'Marketing': 'r', 'Sales': 'g', 'IT': 'b'}
positions = [1, 2, 3]  # use a list instead of np.arange() to avoid the index error
for i, department in enumerate(['Marketing', 'Sales', 'IT']):
    group = df[df['department']==department]
    ax.violinplot(group['salary'], positions=[positions[i]], showmeans=True, showextrema=False, widths=0.6)
    for pc in ax.collections:
        pc.set_facecolor(colors[department])
plt.xticks(positions, ['Marketing', 'Sales', 'IT'])
plt.show()
```

```python
# 48. Create a box plot with separate colors for each group
fig, ax = plt.subplots()
colors = {'Marketing': 'r', 'Sales': 'g', 'IT': 'b'}
data = []
labels = []
for department, group in df.groupby('department'):
    data.append(group['salary'])
    labels.append(department)
bp = ax.boxplot(data, labels=labels, patch_artist=True)
for patch, color in zip(bp['boxes'], [colors[label] for label in labels]):
    patch.set_facecolor(color)
plt.show()
```
